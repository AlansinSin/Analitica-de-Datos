```python
# Importar librerias necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

# Importar libreria para descargar archivo
from google.colab import files
```

```python
#Cargar archivo csv desde equipo
Table = pd.read_csv("CountryTable.csv")
```

```python
# Imprimir las primeras 5 filas
Table.head(5)
```

```python
#Filtro por objeto para los datos con region en el Caribe
filtro1=Table[Table["region"] == "Caribbean"]
filtro1
```

```python
# Descargar archivo de Filtro 1
filtro1.to_csv("Entregable_Filtro1.csv")
files.download("Entregable_Filtro1.csv")
```

```python
#Filtro por filas desde la 10 a la 16
filtro2=Table.iloc[ 10:16 , : ]
filtro2
```

```python
# Descargar archivo de Filtro 2
filtro2.to_csv("Entregable_Filtro2.csv")
files.download("Entregable_Filtro2.csv")
```

```python
#Filtro para solo mostrar columna 1, 4 y 6 (recordando que el conteo empieza en 0)
filtro3=Table.iloc[ : , [1,4,6]]
filtro3
```

```python
# Descargar archivo de Filtro 3
filtro3.to_csv("Entregable_Filtro3.csv")
files.download("Entregable_Filtro3.csv")
```

```python
Table2= pd.read_csv("CountryTable.csv", index_col=2)
Table2
```

```python
#Paso 2: Aplicaf Filtro de filas vs columnas
filtro4=Table2.loc[["North America","Europe"], ["surface_area"]]
filtro4
```

```python
# Descargar archivo de Filtro 4
filtro4.to_csv("Entregable_Filtro4.csv")
files.download("Entregable_Filtro4.csv")
```

```python
#Filtro para mostrar las primeras 10 filas
filtro5=Table2.head(10)
filtro5
```

```python
# Descargar archivo de Filtro 5
filtro5.to_csv("Entregable_Filtro5.csv")
files.download("Entregable_Filtro5.csv")
```

```python
#Filtro para solo mostrar con esperanza de vida mayor a 60
filtro6=Table2[Table2["life_expectancy"] > 60]
filtro6
```

```python
# Descargar archivo de Filtro 6
filtro6.to_csv("Entregable_Filtro6.csv")
files.download("Entregable_Filtro6.csv")
```

```python
#Filtro Y
filtro7=Table2[(Table2["life_expectancy"] > 60) & (Table2["government_form"] == "Republic")]
filtro7
```

```python
# Descargar archivo de Filtro 7
filtro7.to_csv("Entregable_Filtro7.csv")
files.download("Entregable_Filtro7.csv")
```

```python
#Filtro O
filtro8=Table2[(Table2["life_expectancy"] > 60) | (Table2["government_form"] == "Republic")]
filtro8
```

```python
# Descargar archivo de Filtro 8
filtro8.to_csv("Entregable_Filtro8.csv")
files.download("Entregable_Filtro8.csv")
```

```python
#Filtro not
filtro9=Table2[(Table2["life_expectancy"] > 60) & ~(Table2["government_form"] == "Republic")]
filtro9
```

```python
# Descargar archivo de Filtro 9
filtro9.to_csv("Entregable_Filtro9.csv")
files.download("Entregable_Filtro9.csv")
```
